```
def hexToBin(bin):
  char_to_replace = {"f": "0000", "g": "0001", "h": "0010", "i": "0011", "j": "0100",
                     "k": "0101", "l": "0110", "m": "0111", "n": "1000", "o": "1001",
                     "p": "1010", "q": "1011", "r": "1100", "s": "1101", "t": "1110",
                     "u": "1111"}
 
  for key, value in char_to_replace.items():
    bin = bin.replace(key, value)
  
  return bin
 
def binToHex(hex):
  output = ""
  char_to_replace = {"0000": "f", "0001": "g", "0010": "h", "0011": "i", "0100": "j",
                     "0101": "k", "0110": "l", "0111": "m", "1000": "n", "1001": "o",
                     "1010": "p", "1011": "q", "1100": "r", "1101": "s", "1110": "t",
                     "1111": "u"}
 
  for i in range(0, len(hex), 4):
    output = output + hex[i : i + 4]
    for key, value in char_to_replace.items():
      output = output.replace(key, value)
  
  return output
 
def xor(s1, s2):
    r = ''
    for i in range(len(s1)):
        if s1[i] == s2[i]:
            r = r +'0'
        else:
            r = r +'1'  
    return r
 
def inputGenerator():
    i1 = format(random.getrandbits(64), '064b')
    diff = '0100000001011100000000000000000000000100000000000000000000000000'
    i1 = str(i1)
    i2 = xor(i1, diff)
    input1 = f(i1, inverse(ip))
    input2 = f(i2, inverse(ip))
    input1 = ''.join([str(i) for i in input1])
    input2 = ''.join([str(i) for i in input2])
    
    return binToHex(input1) + binToHex(input2)
```

## 6 Round DES Decryption

```
## ith sbox
def sboxi(x,i):
    ## x : 6 bit 
    ## i : no. of xbox
    row = 2 * int(x[0]) + int(x[5])
    col = 8 * int(x[1]) + 4 * int(x[2]) + 2 * int(x[3]) + int(x[4])
    val = sBoxes[i][row][col]
    val = format(val,'04b')
    return val
 
## sbox backward
def sboxAll(x):
    ## x :48 bit string
    val = "" 
    for i in range(8):
        val = val + sboxi(x[i*6:i*6+6],i)
    return val 
```

```
## for decrypt 1 round if
## xor of left and right input is known
## xor of right output is known 
## expansion of right half of both input is known
## output will be set of possible value of key for that round
def decrypt(rinxo, linxo, routxo, E1, E2):
    Sin = f(rinxo, E)
    Pout = np.bitwise_xor(linxo, routxo)
    Sout = f(Pout, inverse(P))
    out = np.zeros(8, dtype=int)
    for i in range(8):
      out[i] = int(''.join(str(i) for i in Sout[i*4: i*4 + 4:1]), 2)
    
    sets = [set() for i in range(8)]
 
    for i in range(0,8):
      for j in range(64):
        E1dash = format(j, '06b')    
        E1dash = np.array(list(E1dash), dtype=int)
        Edash = Sin[i*6:i*6+6:1]
        E2dash = np.bitwise_xor(E1dash,Edash)
 
        S1dash = sboxi(E1dash,i)
        S2dash = sboxi(E2dash,i)
        S1dash = np.array(list(S1dash), dtype=int)
        S2dash = np.array(list(S2dash), dtype=int)
 
        val = int(''.join(str(i) for i in np.bitwise_xor(S1dash, S2dash)), 2)
 
        if val == out[i]:
          key1 = np.bitwise_xor(E1[i*6:i*6+6:1],E1dash)
          key2 = np.bitwise_xor(E2[i*6:i*6+6:1],E2dash)
          k1 = ""
          k2 = ""
          for k in range(6):
            k1 = k1 + str(key1[k])
            k2 = k2 + str(key2[k])
          sets[i].add(k1)
          sets[i].add(k2)
    
    return sets

## solving des right to left if key and output is known (for 1 round)
```

```
def left(key, rout,lout):
  rin = lout # right input = left output
  e = f(rin,E)
  key = np.array(list(key), dtype=int)
  sin = np.bitwise_xor(key,e)
  sout = sboxAll(sin) #string
  sout = np.array(list(sout),dtype=int)
  pout = f(sout,P)
  lin = np.bitwise_xor(pout,rout)
  return lin
```
